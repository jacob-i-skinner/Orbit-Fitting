```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

observationVector = np.array([0, 1]) #we are looking at the ellipse edge on, from "below" the plot

e = 0.8
l = 1
M = 1

#lets make an ellipse
theta = np.linspace(0, 4*np.pi, num=2000)
r     = l/(1+e*np.cos(theta))

fig = plt.subplot(111, projection = 'polar')
fig.plot(theta, r)
```

```python
#this is a plot of the orbital velocity of an object as it sweeps out the orbit plotted above.
V = np.sqrt((M/l)*((e*np.sin(theta))**2+(1+e*np.cos(theta))**2))
plt.plot(theta, V)
plt.xlim(0, 4*np.pi)
```

```python
#convert r and theta into rectangular form
x = np.zeros(theta.size)
y = np.zeros(theta.size)
x = r*np.cos(theta)
y = r*np.sin(theta)

#differentiate r and theta

dx = np.zeros(theta.size)
dy = np.zeros(theta.size)

for i in range(0, theta.size): #below are the velocity vectors, they must be normalized
    dx[i] = x[i] - x[i-1]
    dy[i] = y[i] - y[i-1]
    
#dx = V*(dx/np.sqrt(dx**2+dy**2))
#dy = V*(dy/np.sqrt(dx**2+dy**2))
```

```python
radialVelocity = np.zeros(theta.size)
for i in range(0, theta.size-1):
    radialVelocity[i] = dx[i]*observationVector[0]+dy[i]*observationVector[1]
    

plt.plot(theta, radialVelocity, 'b-')
plt.xlim(0, 4*np.pi)
```
