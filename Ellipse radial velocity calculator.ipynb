```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

#lets make an ellipse

x = np.linspace(-2.5,0.625, num=1000)
y = np.sqrt(((1-0.6*x)**2)-x**2)

plt.grid(True)
plt.plot(x, y)
plt.plot(x, -y)
plt.xlim(-3, 1)
```

```python
theta = np.linspace(0, 2*np.pi, num=1000)
r     = 1/(1+0.6*np.cos(theta))

fig = plt.subplot(111, projection = 'polar')
fig.plot(theta, r)
```
