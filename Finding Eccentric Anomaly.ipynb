```python
P = 1
T = 0
e = 0.9
HJD = np.linspace(0, 100, num = 10000)

rootMeanAnomaly = (2*np.pi/P)*(HJD-T)
rootEccentricAnomaly = rootMeanAnomaly + e*np.sin(rootMeanAnomaly) + ((e**2)*np.sin(2*rootMeanAnomaly)/2)


while True:
    eccentricAnomaly     = rootEccentricAnomaly
    meanAnomaly          = eccentricAnomaly - e*np.sin(eccentricAnomaly)
    rootEccentricAnomaly = eccentricAnomaly +(rootMeanAnomaly-meanAnomaly)/(1-e*np.cos(eccentricAnomaly))
    if np.amax(rootEccentricAnomaly-eccentricAnomaly) < 1E-8:
        break
trueAnomaly = 2*np.arctan(np.sqrt((1 + e)/(1 - e))*np.tan(rootEccentricAnomaly/2))

rv = (np.cos(trueAnomaly) + (e*np.cos(0)))
```
